## Parabolic Mirror Layout

```python
plant = Plant()
n = 5
xs = np.round(np.linspace(plant.x_min, plant.x_max, n + 1))[1:]
ys = list(map(lambda x: np.round( 1/48 * x**2 ), xs))
```

```python
layout = np.column_stack((xs,ys))
plant.layout = layout
plant.set_layout()
print(plant.valid_layout)
print(utils.get_energy(plant, do_stats=False))
plant.draw("parabolic-mirror-layout")
```
